#Looping/Perulangan
Python menyediakan dua statement untuk melakukan proses perulangan for dan while

##For Loop
perulangan for digunakan untuk melakukan iterasi didalam sequence seperti list, tuple, string, dan array atau objek yang dapat diiterasi lainnya.

###For Loop Syntax
    ```
    for val in sequence:
      loop body
    ```

Pada proses iterasi ini, val adalah variabel yang diambil dari elemen elemen di dalam sequence untuk setiap iterasi. Perulangan akan berlangsung hingga elemen terakhir dari sequence. Pada bagian body dari for loop diletakkan dalam indentasi.

###For Loop Flowchart
Screenshot (46).png

````python
#Contoh python for loop
#Program untuk menghitung jumlah semua angka dalam sebuah list

#List angka
list_angka=[6,5,3,8,4,2,5,4,11]

#variabel untuk menyimpan hasil penjumlahan
jumlahan = 0

#iterasi di dalam list
for angka in list_angka:
  jumlahan = jumlahan+angka

print("Jumlah :", jumlahan)
````

###The range() function
Kita dapat membuat sebuah sequence angka mengguunakan fungsi range(). range(10) akan menghasilkan sequence angka dari 0 sampai 9 (10 angka)

Kita juga dapat mendefinisikan nilai awal, nilai akhir dan step sizenya yaitu range(start, stp, step_size). secara default step_size bernilai 1 jika tidak ditentukan secara khusus.

Kita dapat menggunakan fungsi range() di dalam perulangan for untuk melakukan iterasi di dalam sequence numbers. Ini dapat kita kombinasikan dengan fungsi len() untuk melakukan iterasi melalui index dari sebuah sequence.

````python
# program untuk melakukan iterasi di dalam list melalui indeks

musik = ['dangdut','campursari','keroncong']

#iterasi di dalam list melalui indeks
for i in range(len(musik)):
  print("Aku suka", musik[i])
````

###For loop with else
sebuah for loop dapat memiliki opsi blok else. Blok else akan dijalankan jika iterasi telah selesai dilakukan dalam sebuah sequence.

keyword break digunakan untuk menghentikan iterasi. Dalam kasus itu maka bagian else dibatalkan. Sehingga, bagian iterasi else hanya dijalankan jika tidak terjadi break.

````python
# Contoh
digits = [0,1,5]

for i in digits:
  print(i)
else:
  print("Tidak ada elemen yang tersisa")
````

Ketika iterasi berjalan maka akan dilakukan print item hingga iterasi selesai. Ketika iterasipada sequence selesai baru perintah pada blok else dijalankan yaitu print Tidak ada elemen yang tersisa

Statement for...else dapat digunakan bersama dengan keyword break untuk menjalankan blok else hanya ketika break tidak dieksekusi.

````python
# Program untuk menampilkan nilai tugas siswa.

nama_siswa = 'Andri'

daftar_nilai = {'Bambang':90, 'Andri':55, 'Candra':77}

for siswa in daftar_nilai:
  if siswa == nama_siswa:
    print(daftar_nilai[siswa])
    break
else:
  print('Tidak ada data nilai atas nama', nama_siswa, '!')
````

##While Loop
While loop pada Python digunakan untuk melakukan iterasi pada blok kode selama nilai dari test expression adalah true.

Looping ini biasanya digunakan ketika kita tidak tau berapa kali iterasi akan dilakukan hingga sebuah kondisi yang diinginkan dapat dicapai.

###Syntax of while loop in Python
    ```
    while test_expression:
    body of while
    ```

Pada loop while, test expression justru dicheck pertama kali. Kemudian statement - statement di dalam body of while akan dieksekusi jika nilai dari test expression adalah true. Setelah satu iterasi dijalankan, test expression di cek kembali. Proses ini terus berlanjut hingga nilai evaluasi dari test expression adalah false.

Pada Python, bagian body pada loop while dituliskan menggunakan indentasi.

Sebagai catatan tambahan bahwa python menginterpretasikan non-zero values sebagai true. None dan 0 diinterpretasikan sebagai false.

###While Loop Flowchart
Screenshot (48).png

````python
# Contoh
# Program untuk menjumlahkan semua
# bilangan asli hingga suku ke-n
# jumlahan = 1+2+3+...+n

n=10

# nilai awal penjumlahan dan iterasi
jumlahan = 0
i=1

while i<=n:
  jumlahan = jumlahan+i
  i=i+1 #update iterasi

#cetak hasil penjumlahan
print("Jumlahan =", jumlahan)
````

Pada program diatas, test expression akan bernilai True selama nilai iterasi i kurang dari atau sama dengan n.

Kita perli mengupdate nilai dari variabel pencatat iterasi i agar iterasi tidak berlangsung secara terus menerus (infinity loop)

###While loop with else
Sama dengan for loop kita juga dapat mengkombinasikan while loop dengan blok else. Bagian else akan dieksekusi jika niali pada test expression pada while loop bernilai false

````python
# Contoh

counter = 0

while counter < 3:
  print("Di dalam loop")
  counter = counter+1
else:
  print("Di dalam else")
````

Perhatikan bahwa, terdapat nilai tagihan yang negatif. Hentikan perhitungan total_tagihan jika menemukan

####Contoh 3:
Diberikan list_cash_flow kamu diminta untuk melakukan penghitungan total_pemasukan dan
total_pengeluaran menggunakan looping for . Untuk nilai positif berarti adalah pemasukan dan nilai

```
list_cash_flow = [2500000, 5000000, -1000000, -2500000, 5000000, 10000000, -500000
```

````python
list_cash_flow = [
2500000, 5000000, -1000000, -2500000, 5000000, 10000000,
-5000000, 7500000, 10000000, -1500000, 25000000, -2500000
]
total_pengeluaran, total_pemasukan = 0, 0
for dana in list_cash_flow:
  if dana > 0:
    total_pemasukan += dana
  else:
    total_pengeluaran += dana
total_pengeluaran *= -1
print(total_pengeluaran)
print(total_pemasukan)
````
