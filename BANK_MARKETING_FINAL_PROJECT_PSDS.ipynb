## **MANHATTAN DISTANCE FUNCTIONS**

### Pengertian

Fungsi jarak Manhattan merupakan salah satu pengukuran yang paling banyak digunakan meliputi penggantian perbedaan kuadrat dengan menjumlahkan perbedaan absolute (mutlak) dari variabel-variabel. Prosedur ini disebut blok absolute atau lebih dikenal dengan *city block distance*. Fungsi jarak Manhattan pada prinsipnya adalah menghitung jarak secara tegak lurus

image.png

### Formula Manhattan Distance

\begin{array}{rcl} d(x,y) & = & \sum\limits_{i=1}^{n} {|x_i-y_i|)} \\ & = & |x_1-y_1|+|x_2-y_2|+...+|x_n-y_n|\end{array}



### Implementasi Formula Manhattan Distance

Langkah 1 dengan menentukan parameter k (jumlah tetangga paling dekat)

```python
k=3
```

Langkah 2 dengan menghitung kuadrat jarak manhattan objek terhadap data training yang diberikan

```python
def manhattan(x,y):
  return sum(abs(a-b) for a,b in zip(x,y))
```

```python
jarak=[]
for x_i in X:
    jarak_manhatan=manhattan(x_i, x_baru)
    jarak.append(jarak_manhatan)
    print(jarak)
print(jarak)
```

```python
urut=np.sort(jarak)
list_urut=[]
for i in range(len(urut)):
    list_urut.append(urut[i])
print(list_urut)
```

```python
urutan=[]
for i in list_urut:
    urtn=jarak.index(i)
    urutan.append(urtn)
print(urutan)
```

## **MINKOWSKI DISTANCE FUNCTIONS**

### Pengertian

Fungsi jarak Minkowski adalah metrik dalam ruang vektor bernorma yang dapat dianggap sebagai generalisasi dari kedua jarak Euclidean dan jarak Manhattan

### Formula Minkowski Distance

Jarak orde Minkowski ${p}$ (di mana $p$ adalah bilangan bulat) antara dua titik $x=(x_1,x_2,...,x_n)$ dan $y=(y_1,y_2,...,y_n)$ didefinisikan sebagai :

$$d(x,y)=(\sum \limits_{i=1}^{n} {|x_i-y_i|})^{1/r}$$

### Implementasi Minkowski Distance

```python
def akar_ke_n(nilai, akar_n):
    nilai_akar=1/float(akar_n)
    return round(float(nilai)**float(nilai_akar),3)

def minkowski(x,y,p_value):
    return akar_ke_n(sum(pow(abs(a-b), p_value) for a, b in zip(x,y)), p_value)
```

```python
jarak=[]
for x_i in X:
    jarak_minkowski=minkowski(x_i, x_baru,1)
    jarak.append(jarak_minkowski)
    print(jarak)
print(jarak)
```

```python
urut=np.sort(jarak)
list_urut=[]
for i in range(len(urut)):
    list_urut.append(urut[i])
print(list_urut)
```

```python
urutan=[]
for i in list_urut:
    urtn=jarak.index(i)
    urutan.append(urtn)
print(urutan)
```

## **CHEBYCHEV DISTANCE FUNCTIONS**

### Pengertian

Fungsi jarak Chebychev adalah metrik yang jarak antara dua vektornya adalah selisih maksimum di antara sumbu-sumbunya

Jarak Chebyshev antara dua kotak pada papan catur memberikan jumlah gerakan minimal yang diperlukan oleh raja untuk menuju ke sana. Ini disebabkan oleh kemampuan raja untuk bergerak miring/diagonal. Berikut di atas adalah jarak Chebyshev ke tiap kotak dari kotak f3.

### Formula Chebychev Distance
$$d(x,y)=\max\limits_{i=1}^n|x_i-y_i|$$

### Implementasi Chebychev Distance

```python
def chebychev(x,y):
    return max(abs(a-b) for a,b in zip(x,y))
```

## **BRAY-CURTIS DISTANCE FUNCTIONS**

### Pengertian

Bray Curtis Distance disebut juga dengan Sorrensen Distance. Ketika perbedaan dijumlahkan antara variabel yang dinormalkan dengan variabel penjumlahan dari objek, sorrensen distance akan menjadi city block distance yang termodifikasi (Moqhtadaiee & Dempster, 2015).

Sorrensen Distance adalah metode normalisasi jarak yang memandang ruang
sebagai grid. Sorrensen Distance memiliki kemiripan dengan city block distance. Sorrensen Distance memiliki kemampuan yang bagus jika semua nilai koordinatnya positif. Rentang nilainya antara nol dan satu. Normalisasi dilakukan menggunakan perbedaan mutlak dibagi dengan penjumlahan (Kaur, 2014). 

### Formula Bray-Curtis Distance
$$d_{ij}=\sum\limits_{k=1}^{n}\frac{|x_{ik}-x_{jk}|}{(x_{ik}+x_{jk})}$$

### Implementasi Bray Curtis Distance

```python
def bray_curtis(x,y):
  return sum((abs(a-b))/(a+b) for a, b in zip(x,y))
```

```python
jarak=[]
for x_i in X:
    jarak_bray_curtis_distance=bray_curtis(x_i, x_baru)
    jarak.append(jarak_bray_curtis_distance)
    print(jarak)
print(jarak)
```

## **CANBERRA DISTANCE FUNCTIONS**

### Pengertian
Jarak Canberra adalah jumlah nilai perbedaan mutlak dibagi dengan jumlah antara dua variabel

### Formula Canberra Distance :
$$d_{ij}=\sum\limits_{k=1}^n\frac{|x_{ik}-x_{jk}|}{|x_{ik}|-|x_{jk}|}$$

### Implementasi Canberra Distance

## **EUCLIDEAN DISTANCE FUNCTIONS**

### Pengertian

Dalam matematika, jarak Euklides atau metrik Euklides adalah jarak garis lurus "biasa" antara dua titik dalam ruang Euklides. Dengan jarak ini, ruang Euklides menjadi ruang metrik.

### Formula Euclidean Distance

Dalam koordinat Kartesius, jika $ p = (p_1, p_2,..., p_n)$ dan $q = (q_1, q_2,..., q_n)$
adalah dua titik dalam ruang Euklid berdimensi $n$, jarak Euclid (d) dari $p$ ke $q$ (atau sebaliknya) di definisikan dengan rumus Pythagoras seperti berikut.

\begin{array}{rcl} d(x,y) & = & \sum\limits_{i=1}^{n} \sqrt{(x_i-y_i)^2}\\ & = & \sqrt{(x_1 - y_1)^2+(x_2 - x_2)^2+...+(x_n - y_n)^2}
\end{array}


### Implementasi Euclidean Distance

```python
def jarak_euclidean(x,y):
  return np.sqrt(sum(pow(a-b, 2) for a, b in zip(x,y)))
```

```python
jarak=[]
for x_i in X:
    jarak_euclid=jarak_euclidean(x_i, x_baru)
    jarak.append(jarak_euclid)
    print(jarak)
print(jarak)
```

```python
def _knn_euclidean(x):
    # Menghitung jarak antara x baru dengan semua sampel di data latih
    jarak = [dst.euclidean(x, x_latih) for x_latih in X_latih]
    # Mengurutkan data dan mengambil sejumlah k objek
    k_idx = np.argsort(jarak)[: k]
    # Mengambil kelas dari k tetangga terdekat dari data latih
    k_neighbor_labels = [y_latih[i] for i in k_idx]
    # Mengambil kelas yang paling sering muncul
    kelas_paling_banyak = Counter(k_neighbor_labels).most_common(1)
    return kelas_paling_banyak[0][0]

def KNN_euclidean(X_latih, y_latih, k, x_baru):
    y_pred= [_knn_euclidean(x_i) for x_i in x_baru]
    return np.array(y_pred)
```

```python
def bar_plot(df, evalf):

    # Memilih tema plot
    sns.set_theme(style="whitegrid")

    # Membuat sebuah figure dan sebuah axes
    fig, ax=plt.subplots(figsize=(10,8)) # figsize=(x,y) untuk menentukan ukuran gambar

    # Melakukan bar plot
    ax = sns.barplot(data=df,               # Nama DataFrame
                     y=evalf,               # Data sumbu Y
                     x=df.index             # Data sumbu X

                    )
    for bar in ax.patches:
        ax.annotate(format(bar.get_height(), '.2f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=15, xytext=(0, 8),
                   textcoords='offset points')
    # Mengatur sudut rotasi label stepsize sumbu x
    plt.xticks(rotation=80)

    # Mengganti label sumbu x
    ax.set_xlabel('distance')

    # Mengganti label sumbu y
    ax.set_ylabel('{} score'.format(evalf))

    # Mengatur Judul Gambar
    ax.set_title('Evaluasi Skor K-Nearest Neighbor dengan Berbagai Jarak', fontsize=16, fontweight='bold')

    # Menampilkan Gambar
    plt.show()
```
